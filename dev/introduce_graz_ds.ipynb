```python
from grazer_dataset import GrazPedWriDataset
from matplotlib import pyplot as plt
from random import randint

seg_path = '/home/ron/Downloads/raw_segmentations_no_cast.h5'
img_path = '/home/ron/Documents/SemiSAM/data/img_only_front_all_left'
csv_path = '/home/ron/Documents/DenseSeg/dataset/data/graz/dataset_with_cv_split.csv'

ds = GrazPedWriDataset('train', saved_seg_path=seg_path, img_path=img_path, csv_path=csv_path)
```
